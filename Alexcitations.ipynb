```python
# Load required packages
import pandas as pd

# Load the publications data
file_path = r"C:\Users\jawri\Data course\Capstone\CORDIS data\Projects\projectPublications.csv"
publications = pd.read_csv(file_path, sep=";", quoting=3, encoding="utf-8", on_bad_lines="skip")

# Clean up column names by stripping quotes and whitespace, then lowercasing
publications.columns = publications.columns.str.strip().str.replace('"', '').str.lower()

# Now extract valid DOIs
valid_dois = publications['doi'].dropna().astype(str).str.strip()
valid_dois = valid_dois[(valid_dois != '') & (valid_dois != '0')]

# Take first 10
sample_dois = valid_dois.head(10).tolist()
print(sample_dois)

```

```python
import requests

# Clean DOIs
clean_dois = [doi.strip('"') for doi in sample_dois]

# Define base OpenAlex endpoint
base_url = "https://api.openalex.org/works/doi:"

# Query the first DOI as a test
test_doi = clean_dois[0]
response = requests.get(base_url + test_doi)

# Show basic result
if response.status_code == 200:
    data = response.json()
    print(f"Title: {data.get('title')}")
    print(f"Citation count: {data.get('cited_by_count')}")
else:
    print(f"Failed to fetch data for DOI: {test_doi} â€” Status code {response.status_code}")
```
