```python
# Drop duplicate projects
project = project.drop_duplicates(subset='id')

# Convert dates
project['startDate'] = pd.to_datetime(project['startDate'], errors='coerce')
project['endDate'] = pd.to_datetime(project['endDate'], errors='coerce')
project['duration_months'] = (project['endDate'] - project['startDate']).dt.days // 30

# Convert to numeric
project['ecMaxContribution'] = pd.to_numeric(project['ecMaxContribution'], errors='coerce')
project['totalCost'] = pd.to_numeric(project['totalCost'], errors='coerce')

# Drop rows missing key info
project = project.dropna(subset=['id', 'totalCost', 'duration_months'])

# Keep only projects with positive total cost and nonzero duration
project = project[(project['totalCost'] > 0) & (project['duration_months'] > 0)]
```

```python
# keep only the columns needed for building features
# drop rows that are missing key info
# remove exact duplicates
# tidy text formatting in activityType and country
org_trimmed = organization[['projectID', 'organisationID', 'country', 'activityType', 'SME']]
org_trimmed = org_trimmed.dropna(subset=['projectID', 'country', 'activityType'])
org_trimmed = org_trimmed.drop_duplicates()
org_trimmed['activityType'] = org_trimmed['activityType'].str.strip().str.upper()
org_trimmed['country'] = org_trimmed['country'].str.strip().str.upper()
```

```python
# count how many of each partner type (HES, PRC, etc.) are in each project
# reset index so projectID becomes a column again
partner_types = pd.crosstab(org_trimmed['projectID'], org_trimmed['activityType'])
partner_types = partner_types.reset_index()
partner_types.columns = ['projectID'] + [f'partnertype_{col}' for col in partner_types.columns[1:]]
```

```python
# show the first few rows of partner type counts
partner_types.head()

```

```python
# create a new table showing whether each project involves any SME
# each row = one project; has_sme is True if any partner is an SME
sme_flag = org_trimmed.groupby('projectID')['SME'].apply(lambda x: any(x == True)).reset_index()
sme_flag = sme_flag.rename(columns={'SME': 'has_sme'})
sme_flag.head()
```

```python
# Create a new table showing how many partners from each country are involved in each project
country_flags = pd.crosstab(org_trimmed['projectID'], org_trimmed['country']).reset_index()

# Rename columns to add 'country_' prefix (except 'projectID')
country_flags.columns = ['projectID'] + [f"country_{col}" for col in country_flags.columns if col != 'projectID']

# Preview
country_flags.head()
```

```python
# merge partner types, SME flag, and country flags into one table
project_features = partner_types.merge(sme_flag, on='projectID', how='left')
project_features = project_features.merge(country_flags, on='projectID', how='left')

# check result
project_features.head()
```

```python
# Step 1: Split topic paths into Level_1, Level_2, Level_3
euroSciVoc[['Level_1', 'Level_2', 'Level_3']] = euroSciVoc['euroSciVocPath'].str.strip('/').str.split('/', expand=True)[[0, 1, 2]]

# Step 2: Keep one row per project–level combination (we assume each row is already per project-topic pair)
lvl1 = euroSciVoc[['projectID', 'Level_1']].dropna().drop_duplicates()
lvl2 = euroSciVoc[['projectID', 'Level_2']].dropna().drop_duplicates()
lvl3 = euroSciVoc[['projectID', 'Level_3']].dropna().drop_duplicates()

# Step 3: One-hot encode each level
lvl1_encoded = pd.get_dummies(lvl1['Level_1'], prefix='topiclvl1')
lvl1_encoded['projectID'] = lvl1['projectID'].values

lvl2_encoded = pd.get_dummies(lvl2['Level_2'], prefix='topiclvl2')
lvl2_encoded['projectID'] = lvl2['projectID'].values

lvl3_encoded = pd.get_dummies(lvl3['Level_3'], prefix='topiclvl3')
lvl3_encoded['projectID'] = lvl3['projectID'].values

# Step 4: Aggregate back to project-level by grouping and summing
lvl1_flags = lvl1_encoded.groupby('projectID').sum().reset_index()
lvl2_flags = lvl2_encoded.groupby('projectID').sum().reset_index()
lvl3_flags = lvl3_encoded.groupby('projectID').sum().reset_index()

# Step 5: Merge all levels into one table
topic_flags = lvl1_flags.merge(lvl2_flags, on='projectID', how='outer') \
                        .merge(lvl3_flags, on='projectID', how='outer') \
                        .fillna(0)

# ✅ Preview
print(f"✅ Topic columns created: {topic_flags.shape[1] - 1}")
topic_flags.head()

```

```python
# Convert all columns except 'projectID' to integer
for col in topic_flags.columns:
    if col != 'projectID':
        topic_flags[col] = topic_flags[col].astype(int)

# Rename final table
topic_dummies = topic_flags.copy()

# ✅ Confirm shape and types
print("✅ Topic dummies prepared.")
print("📊 Shape:", topic_dummies.shape)
print("🔠 Dtypes:")
print(topic_dummies.dtypes.value_counts())
```

```python
project_features = project_features.merge(topic_dummies, on='projectID', how='left')
```

```python
# convert to numeric, force bad values to NaN
cols = ['totalCost', 'ecMaxContribution', 'duration_months']
for col in cols:
    project[col] = pd.to_numeric(project[col], errors='coerce')

# now drop rows where any of those are missing or not valid
project = project.dropna(subset=cols)

# drop any rows where values are 0 or negative
project = project[
    (project['totalCost'] > 0) &
    (project['ecMaxContribution'] > 0) &
    (project['duration_months'] > 0)
]
```

```python
# Ensure projectID columns are both strings
project_features['projectID'] = project_features['projectID'].astype(str)
project['id'] = project['id'].astype(str)

# Drop existing versions of these columns if they’re already in the table
columns_to_merge = ['duration_months', 'totalCost', 'ecMaxContribution']
project_features = project_features.drop(columns=[col for col in columns_to_merge if col in project_features.columns])

# Perform the merge cleanly
project_features = project_features.merge(
    project.rename(columns={'id': 'projectID'})[['projectID'] + columns_to_merge],
    on='projectID',
    how='left'
)

# Drop rows with missing key project data
project_features = project_features.dropna(subset=columns_to_merge)

print("✅ Metadata merged and cleaned.")


```

```python
# Extract the base project ID from the full ID values like "825664_2581201_PUBL"
projectPublications['projectID'] = projectPublications['projectID'].astype(str).str.extract(r'^(\d+)')

# Keep only unique project–DOI pairs
unique_pubs = projectPublications[['projectID', 'doi']].drop_duplicates()

# Count how many unique publications each project has
pub_counts = unique_pubs.groupby('projectID').size().reset_index(name='num_publications')

# Ensure projectID is str in both tables
pub_counts['projectID'] = pub_counts['projectID'].astype(str)
project_features['projectID'] = project_features['projectID'].astype(str)

# Drop existing 'num_publications' column if it exists (for safe re-runs)
if 'num_publications' in project_features.columns:
    project_features = project_features.drop(columns=['num_publications'])

# Merge publication counts
project_features = project_features.merge(
    pub_counts,
    on='projectID',
    how='left'
)

# Fill missing counts with 0
project_features['num_publications'] = project_features['num_publications'].fillna(0).astype(int)

print("✅ Cleaned and merged publication counts.")
```
